```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('IRIS.csv')
df.head()
```

```python
specie=df['species'].value_counts().sort_index()
plt.bar(specie.index.astype(str),specie.values)
plt.xlabel('Species')
plt.ylabel('Count')
plt.title('Count of Species')
plt.show()
```

```python
plt.hist(df['petal-length'],bins=10,color='green',edgecolor='black')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['petal-length'],df['petal-width'])
plt.title('Relationship between Petal Length and Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```

```python
specie=df['species'].unique()
data=[df[df['species']==s]['petal-length'] for s in specie]
plt.boxplot(data,tick_labels=specie)
plt.title('Boxplot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()
```
