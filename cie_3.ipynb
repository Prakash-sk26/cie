```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('/content/mpg.csv')
df.head()

df.isnull().sum()
df.fillna(method='ffill',inplace=True)
df.isnull().sum()

```

```python
x=df[['hp']]
y=df[['mpg']]
sns.scatterplot(data=df,x='hp',y='mpg')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)

```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
x_predict=lr.predict(x_train)
print('y predict:',y_pred)
print('x predict:',x_predict)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
mae=mean_absolute_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
print('mean scored error:',mae)
print('r2 score:',r2)
print('mean squared error:',mse)

rmse=np.sqrt(mse)
print('root mean squared error:',rmse)
```

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,x_predict,color='black')
plt.title('hp vs mpg')
plt.xlabel('hp')
plt.ylabel('mpg')
plt.show()
```
